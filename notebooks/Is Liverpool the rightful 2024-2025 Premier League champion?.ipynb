```python
import sys
from pathlib import Path
sys.path.append(str(Path().resolve().parent))

import footyIG as fty
import pandas as pd
from tqdm import tqdm
```

```python
print(fty.config.get_all_leagues())
```

```python
score = fty.Scores365()
df = score.get_all_season_games('Premier League')
df
```

```python
team = 'Liverpool'
df_team = df[(df['home_team'] == team) | (df['away_team'] == team)].copy().reset_index(drop = True)
df_team
```

```python
match_stats_list = {}

for i, row in tqdm(df_team.iterrows(), total=len(df_team), desc="Extrayendo partidos"):
    url = row['match_url']
    match_id = row['match_id']
    try:
        df_stats = score.extract_statistics(url)
        match_stats_list[match_id] = df_stats
    except Exception as e:
        print(f"Error en partido {match_id} - {url}: {e}")
```

```python
first_key = list(match_stats_list.keys())[2]
df_example = match_stats_list[first_key]
df_example.head()
```

```python
def normalize_match_stats(match_id, df_stats):
    teams = df_stats.columns[1:]  
    stats = []

    for team in teams:
        team_stats = df_stats[['Statistic', team]].copy()
        team_stats.columns = ['statistic', 'value']
        team_stats['team'] = team
        team_stats['match_id'] = match_id
        stats.append(team_stats)

    return pd.concat(stats, ignore_index=True)
```

```python
normalized_stats = []

for match_id, df in match_stats_list.items():
    try:
        normalized = normalize_match_stats(match_id, df)
        normalized_stats.append(normalized)
    except Exception as e:
        print(f"Error normalizando partido {match_id}: {e}")

df_all_normalized = pd.concat(normalized_stats, ignore_index=True) #new dataframe
```
